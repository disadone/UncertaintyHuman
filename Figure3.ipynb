```python
%matplotlib inline
import utils,os
import importlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyperplot as ppp
from utils import gr_fitted
# from utils import SF

from tqdm import tqdm

import warnings
warnings.filterwarnings("ignore")
mk=ppp.Marker('fig.yml')
aa=np.asarray
```

```python
folder='subj_csv_data'
subjects=['s01','s02','s03','s06','s08','s09','s10','s12','s13','s15']
len_subjects=len(subjects)
```

```python
def reload(no=0):
    global ads,ad,subject,subject_fig,figures_folder,fitted
    importlib.reload(utils)
    ads=[utils.AnalysisData(folder,subject) for subject in subjects]
    ad=ads[no]
    subject=subjects[no]
    subject_fig=subjects[no].split('_')[0]
    figures_folder=f"figs/{subject}"
    if not os.path.exists(figures_folder):
        os.mkdir(figures_folder)
    fitted=np.load(f"connect/data/{subject}.npy",allow_pickle=True).tolist()
```

```python
reload(2)
```

```python
with ppp.SF("%s_proportion_no_opt_correct"%subject,figures_folder,mk) as (fig,axes):
    ad.plot_proportion(ax=axes,data=['no-up lr','opt-up lr','opt-up up'],prop_types=['correct'],remove_up026=False,errorbar=True,dirc=False,comparable=True)
    utils.axSet(axes,xticks=None,
                title="%s proportion under different Coherences"%subject,
                xlabel="|Coherence|",ylabel="Accuracy")
    gr_fitted(axes,fitted["accuracy"])
```

```python
with ppp.SF("%s_responseTimeCoh_no_opt"%subject,figures_folder,mk) as (fig,axes):
    ad.plot_responseTime(ax=axes,data=['no-up lr','opt-up lr','opt-up up'],with_scatter=False,method='errorbar',dirc=True,remove_up026=False)
    utils.axSet(axes,xticks=None,title="%s First Stage Reaction time (s)"%ad.subject,xlabel="Coherence",ylabel="Reaction time (s)")
    gr_fitted(axes,fitted["rt"])
```

```python
with ppp.SF("%s_proportion_no_opt_up"%subject,figures_folder,mk) as (fig,axes):
    ad.plot_proportion(ax=axes,data=['opt-up'],prop_types=['up'],comparable=True,dirc=False,alpha=1,errorbar=True)
    utils.axSet(axes,xticks=None,legend=None,ylim=[0,0.65],title="%s proportion under different Coherences"%ad.subject,xlabel="|Coherence|",ylabel="UR proportion")
    gr_fitted(axes,fitted["ur_prop"])
```
